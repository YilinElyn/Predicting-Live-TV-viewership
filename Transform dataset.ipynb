```python
df_Now_TV['date'] = pd.to_datetime(df_Now_TV['date'], format='%d-%m-%Y')
```

```python
df_Now_TV.head()
```

```python
# Collate data based on time granularity

# Separate minute - based data
minute_columns = [f'videoConsumption[{i}]' for i in range(1440)] + [f'impression[{i}]' for i in range(1440)]
df_minute = df_Now_TV[['date'] + minute_columns]
# Separate day - based data
df_ud = df_Now_TV[['date', 'uniqueDevice']]

# Sort by date
df_minute = df_minute.sort_values(by='date')
df_ud = df_ud.sort_values(by='date')
```

```python
df_minute.head()
```

```python
df_ud.head()
```

```python
# Process videoConsumption and impression data separately
video_columns = [col for col in df_minute.columns if 'videoConsumption' in col]
impression_columns = [col for col in df_minute.columns if 'impression' in col]

# Transform the DataFrame into a long format
df_video = df_minute.melt(id_vars=['date'], value_vars=video_columns, var_name='minute', value_name='videoConsumption')
df_impression = df_minute.melt(id_vars=['date'], value_vars=impression_columns, var_name='minute', value_name='impression')

# Extract minute numbers: extract '1' from a string like 'videoConsumption[1]'
df_video['minute'] = df_video['minute'].str.extract('\[(\d+)\]').astype(int)
df_impression['minute'] = df_impression['minute'].str.extract('\[(\d+)\]').astype(int)

# Merge video and impression data
df_merged = pd.merge(df_video, df_impression, on=['date', 'minute'], how='outer')

# Convert the date to a string format to ensure it can be correctly merged with the time string
df_merged['date'] = df_merged['date'].astype(str)
df_merged['time'] = df_merged['minute'].apply(lambda x: f"{x // 60:02}:{x % 60:02}:00")

# Create a complete datetime column based on the date and minute
df_merged['datetime'] = pd.to_datetime(df_merged['date'] + ' ' + df_merged['time'])

# Drop columns that are no longer needed
df_merged = df_merged.drop(columns=['date', 'minute', 'time'])

# Select and order the final columns
df_minute_new = df_merged[['datetime', 'videoConsumption', 'impression']]

# Sort the DataFrame
df_minute_new = df_minute_new.sort_values(by='datetime')
```
